## Main Techniques

### Array/String
1. given 2 sorted arrays, merge them into 1 sorted array.
    - start with 2 pointers, 1 at the end of each list and compare the values.
1. given a array and you want to remove all the occurrences of an element
    - go through the list and you could create a new list or move elements to the end for faster speed than poping.
1. given a sorted list (asc), remove the duplicates (i.e each element should only appear once).
    - go through the list, use 2 pointer, 1 to point where the next unique number should be stored and 2 to go through the list to check that the number does not = to the number in pos pointer1 - 1.
1. given a sorted list (asc), allow up to 2 occurences of each element.
    - go through each number in the list and check that nums[i-2] is that not number, if not then make nums[i] that number and increase i.
1. Find majority element (more than 50% of the list)
    - 2 ways to do this: 1 - use a dictionary, 2 - Every time current number is the same as res, we add +1 to majority. If not, we add -1 to majority.
1. Rotate array to the right by k steps, 3 ways to do this:
    - 1) pop the last add to front (inefficent)
    - 2) find new position of the number and store number in new array with new position. new position = (old pos + k ) % len nums
    - 3) reverse the nums, then reverse the first k nums and reverse the rest.
1. best time to buy sell stock:
    - keep track of buy price, if price is smaller then buy price is price, calc profit each time and keep track of max_profit.
1. best time to buy sell stock 2:
    - keep track of the lowest price and comapre with price in the list, if price is lower than lower price then lower price = price, if price is > then take profit and lower price = price.
1. Jump Game, we can to find if we can jump to the end:
    - the goal is length - 1, we can jump, pos+num[x], if we can jump to >= goal then new goal is x.
1. Jump Game II (same as 1 but need to find min jumps):
    -  can use recursion to find the jump or, search all possible between near and farthest.
1. h index:

1. o(1) insert, delete and random access:
    - cant just use 1 data structure, need to use 2.
    - use list and dict
    - when removing, move the last element in the list to the index of the val you want to remove then pop, this is now O(1) removal.

    - need to find the prefix product and the suffix product and multiply them
    - have a count c=1, traverse through the list, answer*=c c*=num[x], then set c=1 and reverse through the list and do same.
